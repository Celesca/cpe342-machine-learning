The cost is then computed by summing over all training examples:
$$ J = \frac{1}{m} \sum_{i=1}^m \mathcal{L}(a^{(i)}, y^{(i)})\tag{6}$$

```python
X = np.array(df['X'])
Y = np.array(df['Y'])
```

2. Write the gradient of each coefficients. (3 points)

```python
import random

# Hyperparameters
alpha = 0.01
epoch = 20000 

# Initial coefficients - Random à¸„à¹ˆà¸²
c0 = random.randint(-1,1)
c1 = random.randint(-1,1)
c2 = random.randint(-1,1)

# à¸ˆà¸³à¸™à¸§à¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
n = len(Y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(epoch):

    y_hat = c0 + c1 * np.exp(c2 * X)
    loss = (1/n) * np.sum((Y - y_hat) ** 2)

    dc0 = (-2/n) * np.sum(Y - y_hat)
    dc1 = (-2/n) * np.sum(np.exp(c2 * X) * (Y - y_hat))
    dc2 = (-2/n) * np.sum((Y - y_hat) * c1 * X * np.exp(c2 * X))
    
    # Update coefficients
    c0 -= alpha * dc0
    c1 -= alpha * dc1
    c2 -= alpha * dc2

    if i % 100 == 0:
        print(f"Iteration {i}: Loss = {loss}, c0 = {c0}, c1 = {c1}, c2 = {c2}")

# Final coefficients
print()
print("Final coefficients:")
print("c0 :", c0)
print("c1 :", c1)
print("c2 :", c2)
print("Loss :", loss)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X_line = np.linspace(0, 5, 100)
y_pred = c0 + c1 * np.exp(c2 * X_line)

plt.plot(X_line, y_pred, label='Regression Line', color='blue')
plt.scatter(X, Y, label='Data Points', color='red')
plt.xlabel('X')
plt.ylabel('Y / y_pred')
plt.title('Plot of Regression Line and Data Points')
plt.legend()
plt.show()
```
